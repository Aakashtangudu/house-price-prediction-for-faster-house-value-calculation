```python
import numpy as np
from sklearn.preprocessing import StandardScaler

class LinearRegression:
    def __init__(self, learning_rate=0.1, max_iterations=1000, alpha=0.01, gradient_type="batch"):
        self.learning_rate = learning_rate
        self.max_iterations = max_iterations
        self.alpha = alpha
        self.gradient_type = gradient_type

    def fit(self, X, y):
        # Add a column of ones for the intercept term
        X = np.insert(X, 0, 1, axis=1)
        
        # Initialize the weights
        self.weights = np.zeros(X.shape[1])
        
        # Define the gradient function
        def gradient(X, y, weights):
            y_pred = np.dot(X, weights)
            errors = y_pred - y
            grad = np.dot(X.T, errors) / len(y)
            return grad
        
        # Define the stochastic gradient function
        def stochastic_gradient(X, y, weights):
            i = np.random.randint(len(y))
            xi = X[i]
            yi = y[i]
            y_pred = np.dot(xi, weights)
            error = y_pred - yi
            grad = xi * error
            return grad
        
        # Define the mini-batch gradient function
        def minibatch_gradient(X, y, weights, batch_size=20):
            batch_indices = np.random.choice(len(y), batch_size, replace=False)
            X_batch = X[batch_indices]
            y_batch = y[batch_indices]
            y_pred = np.dot(X_batch, weights)
            errors = y_pred - y_batch
            grad = np.dot(X_batch.T, errors) / len(y_batch)
            return grad
        
        # Select the appropriate gradient function
        if self.gradient_type == "batch":
            grad_func = gradient
        elif self.gradient_type == "stochastic":
            grad_func = stochastic_gradient
        elif self.gradient_type == "mini-batch":
            grad_func = minibatch_gradient
        
        # Perform gradient descent
        for i in range(self.max_iterations):
            grad = grad_func(X, y, self.weights)
            self.weights -= self.learning_rate * (grad + self.alpha * self.weights)
    
    def predict(self, X):
        # Add a column of ones for the intercept term
        X = np.insert(X, 0, 1, axis=1)
        y_pred = np.dot(X, self.weights)
        return y_pred
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

lr = LinearRegression(learning_rate=0.1, max_iterations=1000, alpha=0.01, gradient_type="batch")
lr.fit(X_scaled, y)

# Make predictions on new X values
X_new = np.array([[-122.25,37.85,52,1274,235.0,558,219,5.6431,0,0,0,1,0]])
X_new_scaled = scaler.transform(X_new)
y_pred = lr.predict(X_new_scaled)

print(y_pred)
```
